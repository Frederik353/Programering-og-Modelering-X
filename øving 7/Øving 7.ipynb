# Eksempel 2

I tillegg til å lage grafer fra data fra filer, kan vi se på hvordan bevegelse skjer ved å simulere bevegelse. For eksempel kan vi se på hva som skjer når vi har en gjenstand som beveger seg med jevnt økende akselerasjon. I dette eksempelet lar vi akselerasjonen være $-1,0 m/s^2$, og den øker med $0,5 m/s^2$ for hvert sekund som går. Med andre ord er formelen for akselerasjonen $a(t)=0,5t-1$. (Dette er bare en tilfeldig valgt likning.)

Vi vil la tida gå framover gradvis, i små skritt. Akselerasjonen vil oppdateres etter likninga over, farta vil være gitt ved farta i forrige tidssteg pluss fartsøkningen i dette tidssteget; $v_1 = v_0 + a_0 \cdot \Delta t$. Posisjonen vil defineres på samme måte. Vi starter med $v = 0$ og $s = 0$.

```python
def aks(t):
# akselerasjonen
    return 0.5*t - 1

n = 1000 # antall målepunkter
start = 0
slutt = 8
dt = (slutt - start)/n

#startverdier
a = aks(0)
v = 0 
s = 0

#lister med verdier som fylles underveis
tider = np.linspace(start, slutt, n)
posisjoner = np.zeros(n)
farter = np.zeros(n)
akselerasjoner = np.zeros(n)

#simulering
for i in range(n):
    a = aks(tider[i])
    v = v + a*dt
    s = s + v*dt
    akselerasjoner[i] = a
    farter[i] = v
    posisjoner[i] = s
    
    
plt.grid()
plt.title("Posisjon-, fart- og akselerasjonsgraf ")
plt.xlabel("Tid (s)")
plt.ylabel("Posisjon (m) / Fart (m/s) / Akselerasjon (m/s^2)")
plt.plot(tider,posisjoner, label = "Posisjon (m)")
plt.plot(tider,farter, label = "Fart (m/s)")
plt.plot(tider,akselerasjoner, label = "Akselerasjon (m/s^2)")
plt.legend()
plt.show()

print(f"Etter {slutt:.2f}s er akselerasjonen {a:.2f}m/s^2, farten {v:.2f}m/s og posisjonen {s:.2f}m.")
```

Med utgangspunkt i et slikt simuleringsprogram er det mulig å gjøre mange fysikksimuleringer, men man må selvfølgelig vurdere hva som må endres for at programmet skal fungere optimalt.
