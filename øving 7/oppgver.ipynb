# Eksempel 2

I tillegg til å lage grafer fra data fra filer, kan vi se på hvordan bevegelse skjer ved å simulere bevegelse. For eksempel kan vi se på hva som skjer når vi har en gjenstand som beveger seg med jevnt økende akselerasjon. I dette eksempelet lar vi akselerasjonen være $-1,0 m/s^2$, og den øker med $0,5 m/s^2$ for hvert sekund som går. Med andre ord er formelen for akselerasjonen $a(t)=0,5t-1$. (Dette er bare en tilfeldig valgt likning.)

Vi vil la tida gå framover gradvis, i små skritt. Akselerasjonen vil oppdateres etter likninga over, farta vil være gitt ved farta i forrige tidssteg pluss fartsøkningen i dette tidssteget; $v_1 = v_0 + a_0 \cdot \Delta t$. Posisjonen vil defineres på samme måte. Vi starter med $v = 0$ og $s = 0$.

```python
def aks(t):
# akselerasjonen
    return 0.5*t - 1

n = 1000 # antall målepunkter
start = 0
slutt = 8
dt = (slutt - start)/n

#startverdier
a = aks(0)
v = 0 
s = 0

#lister med verdier som fylles underveis
tider = np.linspace(start, slutt, n)
posisjoner = np.zeros(n)
farter = np.zeros(n)
akselerasjoner = np.zeros(n)

#simulering
for i in range(n):
    a = aks(tider[i])
    v = v + a*dt
    s = s + v*dt
    akselerasjoner[i] = a
    farter[i] = v
    posisjoner[i] = s
    
    
plt.grid()
plt.title("Posisjon-, fart- og akselerasjonsgraf ")
plt.xlabel("Tid (s)")
plt.ylabel("Posisjon (m) / Fart (m/s) / Akselerasjon (m/s^2)")
plt.plot(tider,posisjoner, label = "Posisjon (m)")
plt.plot(tider,farter, label = "Fart (m/s)")
plt.plot(tider,akselerasjoner, label = "Akselerasjon (m/s^2)")
plt.legend()
plt.show()

print(f"Etter {slutt:.2f}s er akselerasjonen {a:.2f}m/s^2, farten {v:.2f}m/s og posisjonen {s:.2f}m.")
```

Med utgangspunkt i et slikt simuleringsprogram er det mulig å gjøre mange fysikksimuleringer, men man må selvfølgelig vurdere hva som må endres for at programmet skal fungere optimalt.

Lag et program som finner arealet under forskjellige funksjoner. Finn arealet under minst 5 forskjellige funksjoner i tillegg til $f(x) = x$, $g(x) = e^x$ og $h(x) = \ln(x)$. Velg grensene selv (a og b i figuren).


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate

# def f(x): return 1/(1+x**2)
# def f(x): return np.sin(x)
# def f(x): return 2*x**2 +4*x +8
def f(x): return np.sin(7*x) # med denne er det lett å se hvor mye bedre simpsons-integrasjon  er i forhold til de andre spessielt for å se forskjell på trapes metoden og simpsons-integrasjon 

# funksjoner spesifisert i oppgaven
# def f(x): return x
# def f(x): return np.e * x
# def f(x): return np.ln(x)
```

```python
a = 0
b = 2*np.pi
N = 10
n = 100  # Use n*N+1 points to plot the function smoothly
dx = (b - a)/N

x = np.linspace(a, b, N+1)
y = f(x)

X = np.linspace(a, b, n*N+1)
Y = f(X)


# finner integral å teste opp mot
area = integrate.quad(f, a, b)
print(f" area: { round(area[0],5) }, error: {area[1] }")
```

```python

# summerer array fra riemansummene
def sum_area(y_vals):
    return round(np.sum(y_vals) * N / (b - a), 2)

# har laget en funksjon for å lage grafene siden dette gjør det lettere å se hva som gjøres forskjellig mellom de forskjellige metodene, hindrer unødvenig mye repetert kode og øker lesbarheten.

def sum_under(plot, graphX, graphY, barX, barY):
    plt.figure(figsize=(30, 10))
    # plt.grid()
    plt.plot(graphX, graphY, 'b')  # plotter graf
    # plotter punk hvor høyden på bokesne blir bestemt
    plt.plot(barX, barY, 'b.', markersize=10)
    # plotter bokser
    if plot == 1:  # left
        plt.bar(barX, barY, width=dx, alpha=0.2,
                align='edge', edgecolor='b')
    elif plot == 2:  # midpoint
        plt.bar(barX, barY, width=dx, alpha=0.2, edgecolor='b')
    elif plot == 3:  # right
        plt.bar(barX, barY, width=-dx, alpha=0.2,
                align='edge', edgecolor='b')
    return sum_area(barY)

# funksjon for å finne øvre og nedre riemannsummer, trenger egen funksjon siden man må sjekke hvilket av de to punktene y er høyest/ lavest i
# tar in bool upper hvis true sjekk øvre riemann sum hvis false sjekk nedre
def riemann_sum(upper):
    sum_array = []
    prev = x[0]
    for i in x[1:]:
        if (f(i) > f(prev)) == upper:  # XNOR gate
            plt.bar(i, f(i), width=-(b-a)/N,
                    alpha=0.2, align='edge', edgecolor='b', color="tab:blue")
            plt.plot(i, f(i), 'b.', markersize=10)
            sum_array.append(f(i))
        else:
            plt.bar(prev, f(prev), width=(b-a)/N,
                    alpha=0.2, align='edge', edgecolor='b', color="tab:blue")
            plt.plot(prev, f(prev), 'b.', markersize=10)
            sum_array.append(f(prev))
        prev = i
    return sum_area(sum_array)
```

```python

# Left
x_left = x[:-1]
y_left = y[:-1]
sum = sum_under(1, X, Y, x_left, y_left)
plt.title(f"Left Riemann Sum, N = {N}, A = {sum}")
plt.show()
```

```python

# Midpoint
x_mid = (x[:-1] + x[1:])/2
y_mid = f(x_mid)
sum = sum_under(2, X, Y, x_mid, y_mid)
plt.title(f"Midpoint Riemann Sum, N = {N}, A = {sum}")

plt.show()
```

```python

# Right
x_right = x[1:]
y_right = y[1:]
sum = sum_under(3, X, Y, x_right, y_right)
plt.title(f"Right Riemann Sum, N = {N}, A = {sum}")
plt.show()
```

```python

# lower
sum_under(4, X, Y, x_right, y_right)
sum = riemann_sum(False)
plt.title(f"Lower Riemann Sum, N = {N}, A = {sum}")
plt.show()
```

```python

# upper
sum_under(5, X, Y, x_right, y_right)
sum = riemann_sum(True)
plt.title(f"Upper Riemann Sum, N = {N}, A = {sum}")
plt.show()
```

```python
# simpsons rule
def p(x, a, b, c): return (a*(x**2))+(b*x)+c


sum_under(7, X, Y, [], [])


def simps(x, y):
    print(N)
    if N % 2 == 1:
        raise ValueError("N must be an even integer.")
    Area = dx/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])

    points = np.column_stack((x, y)) # setter sammen x og y kordinatene til en array med form (n,2) altså array med arrayer hvor index 0 er x koordinatet og 1 er y koordinatet til ett punkt
    prev = points[0] # starter ved å sette forige punk som første punkt
    for current in points[1:]: #looper over alle punkter bortsett fra første punkt
        # finner tre punkter
        p1 = prev
        midpointX = current[0] - (dx/2)
        p2 = [midpointX, f(midpointX)]
        p3 = current
        a, b, c = calc_parabola_vertex(p1, p2, p3) # finner faktorene a b og c i en andregradsformel ut fra punktene

        x_pos = np.linspace(current[0],  prev[0], n+1)
        y_pos = p(x_pos, a, b, c)

        plt.plot(x_pos, y_pos, "b")  # plotter grafen mellom punktene
        plt.fill_between(x_pos, y_pos, color="tab:blue", alpha=0.2) # fyller areal under graf for å gjøre det lettere å se
        # plt.scatter(p1[0], p1[1], color='r', marker="D", s=50)  # 1st known xy
        # plt.scatter(p2[0], p2[1], color='g', marker="D", s=50)  # 2nd known xy
        plt.plot(p2[0], p2[1], 'b.', markersize=10)
        prev = current

    return Area


def calc_parabola_vertex(p1, p2, p3): # klar over at dette er så å si uleselig
    denom = (p1[0]-p2[0]) * (p1[0]-p3[0]) * (p2[0]-p3[0])
    A = (p3[0] * (p2[1]-p1[1]) + p2[0] *
         (p1[1]-p3[1]) + p1[0] * (p3[1]-p2[1])) / denom
    B = ((p3[0]**2) * (p1[1]-p2[1]) + (p2[0]**2) *
         (p3[1]-p1[1]) + (p1[0]**2) * (p2[1]-p3[1])) / denom
    C = (p2[0] * p3[0] * (p2[0]-p3[0]) * p1[1]+p3[0] * p1[0] *
         (p3[0]-p1[0]) * p2[1]+p1[0] * p2[0] * (p1[0]-p2[0]) * p3[1]) / denom
    return A, B, C


sum = round(simps(x, y), 3) # runder av areal

plt.title(f"Simpsons rule, N = {N}, A = {sum}")

plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
import math

k = 0.47 # avkjølings konstant = alpha * A / C hvor alpha er varmeoverføringskoeffisienten, A er overflatearealet og C er varmekapasiteten, siden disse i vårt tilfelle er konstante gjennom målingen kan dette skrives som en konstat k
tEnv = -15 # temperatur til omgivelsene
startT = 27 # temp til måler i starten
# microbitten sier det var 27 grader i starten noe jeg tviler litt på at stemmer
# dette kan komme av at mikrobitten bruker en nRF52833 prosessor hvor temperatur måleren ligger rett ved sidenav prosessoren og er ment til å måle temperaturen til denne ikke omgivelsene men i følge microbit skal dette gi en "ganske god anslåelse av temperaturen", en egen sensor for temperatur ville fortsatt gitt mer nøyaktige målinger og ikke bli påvirket av cpu temperaturen.

y = [startT]
x = np.linspace(0, 10, 10*60) # like mange punkter som i målingen

```

bruker newtons kjølingslov

dQ/dt=-alpha * A * (Ts - T)

Q = C*T

dT/dt = alpha * A / C * (Ts) = k * (Ts - T)

løser denne diferensialligningen og får 

T(t) = Ts + (T0 * Ts)e^-k*t

dette gjør vi så om til en funksjon  under


```python

for dt, val in enumerate(x[1:]):
    y.append(tEnv + (startT - tEnv) * math.e**(-k * val))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
var = pd.read_excel("./UN_world_population_data.xlsx", header=16) # åpner fil med data fra FN

x = list(var.loc[: , "1950" : "2100"]) # leser data til x akse, dataen var skrevet horisonttalt som gjorde alt litt verre
x = list(map(int, x)) # cellene leses automatisk som string og må gøres om til int hvis ikke vil aksene bli fylt opp med verdier
un_y = var.iloc[0, 7 : 160] # samme for populasjonsverdiene
un_y = list(map(int, un_y))
```

```python
def logistic_model(t_0, t_end,L,k, p_0, forkyvining): # funksjon for logistisk model 
    results = [p_0]
    a = 0
    b = a + 150
    for t in range(a, b):
        results.append((L) / (1 + forkyvining* math.e **(-k *(t))))
    return results
```

```python
# L =  10875394 # dette er det Fn anslår som befolningstall i år 2100, kan ikke brukes som maks populasjon siden vi trolig fortsatt ikke har nådd maks populasjon innen 2100 noe som reflekteres i dataen fra FN
L = 11_500_000 # anslått maks populasjon
k = 0.03 # vokst konstant
forskyvning = 3.8 # forskyver grafen horisonttalt siden befolknings dataen muligens er et annet sted i perioden og vi må derfor forskyve modellen for at den skal passe

model_y = logistic_model(x[0], 2100 ,  L, k, un_y[0], forskyvning)
```

```python

# plotter grafene
n = np.linspace(0,10,10)
fig, ax = plt.subplots(figsize=(15, 8))

plt.plot(x, model_y, c="b", label="Model Prediction")
plt.plot(x[:72], un_y[:72], c="red", label="UN Data", linewidth=1) # endrer linjestil for å se forskjell på FN sine data og det det antar i fremtiden
plt.plot(x[71:], un_y[71:], c="chocolate", linestyle ="dashed", label="UN prediction", linewidth=1)
ax.ticklabel_format(style='plain')
plt.legend(loc="best")
plt.xlabel("Year")
plt.ylabel("Population")
plt.locator_params(axis="x", nbins=20)
plt.locator_params(axis="y", nbins=10)
plt.grid()
plt.title("Logistic World Population model")
plt.show()
```
